```python
# Code for Question 4C

import numpy as np

matrices = [np.random.randint(-10, 11, size=(3, 3)) for _ in range(3)]

for i, A in enumerate(matrices, start=1):
    print(f"\nMatrix A{i}:")
    print(A)

    try:
        A_inv = np.linalg.inv(A)
        print(f"\nInverse of A{i}:")
        print(A_inv)

        print(f"Product:")
        print(np.round(A @ A_inv, 6))

    except np.linalg.LinAlgError:
        print(f"\nMatrix A{i} is not invertible.")
```

```python
# Setup for Problem-5
import pandas as pd

df = pd.read_csv("kc_house_data.csv")
df = df.drop(columns=["id", "date", "zipcode"])

# Separating both variables
y = df["price"]
X = df.drop(columns=["price"])
```

```python
statistics = pd.DataFrame({
    "mean": X.mean(),
    "min": X.min(),
    "max": X.max(),
    "variance": X.var(),
})
print(statistics)
```

```python
lowest_avg = statistics["variance"].idxmin(), statistics["variance"].min()
highest_avg = statistics["variance"].idxmax(), statistics["variance"].max()

print("Lowest average:", lowest_avg)
print("Highest average:", highest_avg)
```

```python
correlations = X.corrwith(y).sort_values(ascending=False)

corr_table = pd.DataFrame({
    "Feature": correlations.index,
    "Correlation with Price": correlations.values
})

print(corr_table)
```
