```python
# Code for Question 4C

import numpy as np

matrices = [np.random.randint(-10, 11, size=(3, 3)) for _ in range(3)]

for i, A in enumerate(matrices, start=1):
    print(f"\nMatrix A{i}:")
    print(A)

    try:
        A_inv = np.linalg.inv(A)
        print(f"\nInverse of A{i}:")
        print(A_inv)

        print(f"Product:")
        print(np.round(A @ A_inv, 6))

    except np.linalg.LinAlgError:
        print(f"\nMatrix A{i} is not invertible.")
```
