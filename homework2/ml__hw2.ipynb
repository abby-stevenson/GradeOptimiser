```python
'''QUESTION 2A'''
import pandas as pd
import numpy as np

train = pd.read_csv('train.csv')
print(train.dtypes)

non_numerical = ['ID'] + ['Gender'] + ['Occupation'] + ['BMI_Category'] + ['Blood_Pressure'] + ['Sleep_Disorder']
train_v2 = train.drop(columns=non_numerical)

train_v2.head()
```

```python
"""QUESTION 2B"""
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

X_train = train_v2.drop('Sleep_Quality', axis=1)
y_train = train_v2['Sleep_Quality']

model = LinearRegression()
model.fit(X_train, y_train)
print("Intercept: ", model.intercept_)
print("Coefficients: ")
for feature, coeff in zip(X_train.columns, model.coef_):
    print(feature, " : ", coeff)

print()

y_pred_train = model.predict(X_train)
training_mse = mean_squared_error(y_train, y_pred_train)
print("Mean Squared Error: ", training_mse)
training_r2 = r2_score(y_train, y_pred_train)
print("R-Squared Value: ", training_r2)
```

```python
"""QUESTION 2C"""

test = pd.read_csv('test.csv')
test_v2 = test.drop(columns=non_numerical)

X_test = test_v2.drop('Sleep_Quality', axis=1)
y_test = test_v2['Sleep_Quality']

y_pred_test = model.predict(X_test)
test_mse = mean_squared_error(y_test, y_pred_test)
print("Mean Squared Error: ", test_mse)
test_r2 = r2_score(y_test, y_pred_test)
print("R-Squared Value: ", test_r2)
```

```python
"""QUESTION 2D"""
for column in non_numerical:
    if column == 'ID':
        continue
    print(column, ":", train[column].nunique())

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

remove = ['ID'] + ['Occupation'] + ['BMI_Category'] + ['Blood_Pressure']
train_v3 = train.drop(columns=remove)
test_v3 = test.drop(columns=remove)

train_v3['Gender_Encoding'] = train_v3['Gender'].map({ 'Female': 0, 'Male': 1 })
test_v3['Gender_Encoding'] = test_v3['Gender'].map({ 'Female': 0, 'Male': 1 })
train_v3.drop('Gender', axis=1, inplace=True)
test_v3.drop('Gender', axis=1, inplace=True)

train_v3['Sleep_Disorder'] = train_v3['Sleep_Disorder'].fillna('None')
test_v3['Sleep_Disorder'] = test_v3['Sleep_Disorder'].fillna('None')

train_sleep = pd.get_dummies(train_v3['Sleep_Disorder'], prefix='Sleep_Disorder', drop_first=True)
test_sleep = pd.get_dummies(test_v3['Sleep_Disorder'], prefix='Sleep_Disorder', drop_first=True)

train_v3 = pd.concat([train_v3, train_sleep], axis=1)
test_v3 = pd.concat([test_v3, test_sleep], axis=1)

train_v3.drop('Sleep_Disorder', axis=1, inplace=True)
test_v3.drop('Sleep_Disorder', axis=1, inplace=True)

y_train = train_v3['Sleep_Quality']
X_train = train_v3.drop('Sleep_Quality', axis=1)
y_test = test_v3['Sleep_Quality']
X_test = test_v3.drop('Sleep_Quality', axis=1)

model_improved = LinearRegression()
model_improved.fit(X_train, y_train)
print("TRAINING VALUES: <><><><><><><><><><><><<><><>")
print("Intercept: ", model_improved.intercept_)
print("Coefficients: ")
for feature, coeff in zip(X_train.columns, model_improved.coef_):
    print(feature, " : ", coeff)
print()

y_pred_train = model_improved.predict(X_train)
training_mse = mean_squared_error(y_train, y_pred_train)
print("Mean Squared Error: ", training_mse)
training_r2 = r2_score(y_train, y_pred_train)
print("R-Squared Value: ", training_r2)

print("TESTING VALUES: <><><><><><><><><><><><<><><>")
y_pred_test = model_improved.predict(X_test)
test_mse = mean_squared_error(y_test, y_pred_test)
print("Mean Squared Error: ", test_mse)
test_r2 = r2_score(y_test, y_pred_test)
print("R-Squared Value: ", test_r2)
```

```python
"""QUESTION 3A"""
def closed_form_lin_reg(X, y):
    x_term = (X.T @ X)
    inverse_term = np.linalg.inv(x_term)
    normal_term = X.T @ y
    theta = inverse_term @ normal_term
    return theta

train_v4 = pd.read_csv('train.csv')
test_v4 = pd.read_csv('test.csv')

non_numerical = ['ID'] + ['Gender'] + ['Occupation'] + ['BMI_Category'] + ['Blood_Pressure'] + ['Sleep_Disorder']
train_v4 = train_v4.drop(columns=non_numerical)
test_v4 = test_v4.drop(columns=non_numerical)

X_train = train_v4.drop('Sleep_Quality', axis=1)
X_test = test_v4.drop('Sleep_Quality', axis=1)
y_train = train_v4['Sleep_Quality']
y_test = test_v4['Sleep_Quality']

X_train = np.c_[np.ones(X_train.shape[0]), X_train]
X_test = np.c_[np.ones(X_test.shape[0]), X_test]

theta = closed_form_lin_reg(X_train, y_train)
print("Intercept:", theta[0])
feature_names = list(train_v4.drop('Sleep_Quality', axis=1).columns)
for i, (name, coeff) in enumerate(zip(feature_names, theta[1:]), 1):
    print(name, ":", coeff)

def predict(X, theta):
    return X @ theta
```

```python
"""QUESTION 3B"""
print("TRAINING VALUES: <><><><><><><><><><><><<><><>")
y_pred_train = predict(X_train, theta)
training_mse = mean_squared_error(y_train, y_pred_train)
print("Mean Squared Error: ", training_mse)
training_r2 = r2_score(y_train, y_pred_train)
print("R-Squared Value: ", training_r2)

print("TESTING VALUES: <><><><><><><><><><><><<><><>")
y_pred_test = predict(X_test, theta)
test_mse = mean_squared_error(y_test, y_pred_test)
print("Mean Squared Error: ", test_mse)
test_r2 = r2_score(y_test, y_pred_test)
print("R-Squared Value: ", test_r2)
```
