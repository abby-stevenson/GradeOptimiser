```python
'''QUESTION 2A'''
import pandas as pd
import numpy as np

train = pd.read_csv('train.csv')
print(train.dtypes)

non_numerical = ['ID'] + ['Gender'] + ['Occupation'] + ['BMI_Category'] + ['Blood_Pressure'] + ['Sleep_Disorder']
train_v2 = train.drop(columns=non_numerical)

train_v2.head()
```

```python
"""QUESTION 2B"""
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

X_train = train_v2.drop('Sleep_Quality', axis=1)
y_train = train_v2['Sleep_Quality']

model = LinearRegression()
model.fit(X_train, y_train)
print("Intercept: ", model.intercept_)
print("Coefficients: ")
for feature, coeff in zip(X_train.columns, model.coef_):
    print(feature, " : ", coeff)

print()

y_pred_train = model.predict(X_train)
training_mse = mean_squared_error(y_train, y_pred_train)
print("Mean Squared Error: ", training_mse)
training_r2 = r2_score(y_train, y_pred_train)
print("R-Squared Value: ", training_r2)
```

```python
"""QUESTION 2C"""

test = pd.read_csv('test.csv')
test_v2 = test.drop(columns=non_numerical)

X_test = test_v2.drop('Sleep_Quality', axis=1)
y_test = test_v2['Sleep_Quality']

y_pred_test = model.predict(X_test)
test_mse = mean_squared_error(y_test, y_pred_test)
print("Mean Squared Error: ", test_mse)
test_r2 = r2_score(y_test, y_pred_test)
print("R-Squared Value: ", test_r2)
```
